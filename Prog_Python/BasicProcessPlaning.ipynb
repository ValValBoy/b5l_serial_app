## ① Check connections

Ici on s'assure que COM4 est bien connectée, c'est sur cette sortie qu'est branchée la B5L. Si COM4 apparait c'est qu'elle est correctement branchée et qu'elle est prête à échanger des informations. 

```python
def serial_ports():
    
    if sys.platform.startswith('win'):
        ports = ['COM%s' % (i + 1) for i in range(256)]
    
    result = []
    for port in ports:
        try:
            s = serial.Serial(port)
            s.close()
            result.append(port)
        except serial.SerialException as e:
            if e.errno == 13:
                raise e
            pass
        except OSError:
            pass
    return result


print(serial_ports()) 
```

On engage ici la connection avec la Caméra

```python
serialPort = serial.Serial('COM4', baudrate=115200)   #Baud à définir
```

GET VERSION

```python
hexadecimal_string = "FE 00 00 00" # Get version
byte_array = bytearray.fromhex(hexadecimal_string)
print('Ping : ', byte_array)

serialPort.write(byte_array)

time.sleep(0.5)

data = serialPort.read_all()

x = binascii.hexlify(data)
y = str(x,'ascii')


print('Normal setup  FE 00 00 00 00 +DATA ;; ', y)
```
